# Iris Flower Classification
---
## Objective

The aim is to classify iris flowers from three species by using length and width measurements of sepals and petals. The iris data set contains 3 classes of 50 instances where each class is a species. The main goal is to design a model that makes a useful classification when provided a new flowers data. The models that will be tested are SVM, KNN, Naive Bayes, Decision Tree Classifier, and Logistic Regression.

**Label Encoding**

Encode the categorical target variable so it can be processed in our Machine Learning Model.

```python
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
iris_df["species_encoded"] = le.fit_transform(iris_df["species"])
iris_df["species_encoded"].unique()
```

**Splitting the Target and Predictor variables**

Once the data has been properly encoded we will split the data into X and y for testing. "y" will be our target variable and "X" will be our predictor variable(s).

```python
X = iris_df.drop(["species", "species_encoded"], axis=1)
y = iris_df["species_encoded"]
```

```python
# Make sure the X and y have the same number of observations for our model
print("N(X):", len(X), "N(y):", len(y))
```

**Highly Correlated Featers**

It is important to address Multicollinearity for models like Linear Regression, Logistic Regression, and KNN. It creates unstable coefficients in regression models. Leads to overfitting in our model. Makes it difficult for feature importance analysis. We can also reduce Dimensionality if necessary but in this case it is not a problem.

Addressing Correlated variables:
If there are several correlated variables you can combine them into a new feature using something like PCA (Princiapl Component Analysis) or Feature Ratios.

Correlation can help with Model Selection:
High correlation -> Use decision tress.
Low correlation -> Linear models work fine.

```python
plt.figure(figsize=(8,6))
sns.heatmap(iris_df.drop(["species"], axis=1).corr(), annot=True, cmap="RdBu", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Model Development
---

**Split the DataFrame into train and test set**

We have to split out data so that we can test our model on new data to prevent overfitting.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

**Feature Scaling**

```python
# Apply Standardization
standard_scaler = StandardScaler()
X_train_scaled = standard_scaler.fit_transform(X_train)
X_test_scaled = standard_scaler.transform(X_test)
```

The Standard Scaler was used which is Z-score Scaling which is bets when data is normally distributed. Best used for KNN, SVM, and Logistic Regression. KNN though is often preferred to MinMaxScaler because the values range between 0 and 1 which make euclidean distance more comparable. We will test the two standardization procedures to find the best one for performance.

### Decision Tree Classifier

```python
# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Instantiate the DecisionTree Classifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train) # Fit the classifier on train data

# Make predictions
y_pred = dt_model.predict(X_test)

# Measure accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy:.3f}")

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Convert confusion matrix to DataFrame for readibility
cm_df = pd.DataFrame(cm, index=dt_model.classes_, columns=dt_model.classes_)

#Visualize confusion matrix using seaborn heatmap
plt.figure(figsize=(6,5))
sns.heatmap(cm_df, annot=True, fmt="d", cmap="Blues", cbar=False, linewidths=0.5)
plt.title("Confusion Matrix - Decision Tree")
plt.ylabel("Actual Species")
plt.xlabel("Predicted Species")
plt.show()
```

**Visualize Decision Tree**

Let's see how the Decision Tree Structure looks.

```python
from sklearn.tree import plot_tree

# Set up plot
plt.figure(figsize=(16, 10))  # Adjust size as needed

# Plot the tree
plot_tree(
    dt_model,
    feature_names=X.columns,     # Show feature names
    class_names=le.classes_,  # Show species names
    filled=True,                 # Color the nodes by class
    rounded=True,                # Rounded boxes
    fontsize=10                  # Adjust text size
)

plt.title("Decision Tree Visualization")
plt.show()
```

**Decision Boundary**

How is the decision tree making decision? Let's take a look using meshgrid at how it is working under the hood. We have to use 2 features as it works in two dimensions. In this case petal_length_cm and petal_width_cm, since they show strong separation between species. 

*We are simplifying the model on for the visualization purpose.*

```python
from matplotlib.colors import ListedColormap
import matplotlib.cm as cm
import numpy as np
import matplotlib.pyplot as plt

# Get 3 distinct colors from viridis
viridis= cm.colormaps['viridis']
viridis_colors = [viridis_cmap(i / 3) for i in range(3)]  # Divide by 3 for evenly spaced colors
viridis_listed = ListedColormap(viridis_colors)

def plot_decision_boundary(X_set, y_set, model, title, label_encoder=None, cmap_colors=viridis_listed):
    X1 = X_set.iloc[:, 0].values
    X2 = X_set.iloc[:, 1].values

    # Meshgrid
    x1_min, x1_max = X1.min() - 1, X1.max() + 1
    x2_min, x2_max = X2.min() - 1, X2.max() + 1
    xx1, xx2 = np.meshgrid(
        np.arange(x1_min, x1_max, 0.01),
        np.arange(x2_min, x2_max, 0.01)
    )
    grid_df = pd.DataFrame(np.array([xx1.ravel(), xx2.ravel()]).T, columns=X_set.columns)
    Z = model.predict(grid_df).reshape(xx1.shape)

    # Plotting decision regions
    plt.figure(figsize=(8, 6))
    plt.contourf(xx1, xx2, Z, alpha=0.5, cmap=ListedColormap(cmap_colors))
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # Plotting data points
    for idx, cls in enumerate(np.unique(y_set)):
        class_label = label_encoder.inverse_transform([cls])[0] if label_encoder else cls
        plt.scatter(X1[y_set == cls], X2[y_set == cls],
                    c=[viridis_listed(idx)], label=class_label, edgecolor='k', s=60)

    plt.title(title)
    plt.xlabel(X_set.columns[0])
    plt.ylabel(X_set.columns[1])
    plt.legend(title="Species")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# 2-feature set for visualization
visual_features = ["petal_length_cm", "petal_width_cm"]
X_vis = iris_df[visual_features]

# Split for visual model (must use same y)
X_train_vis, X_test_vis, _, _ = train_test_split(
    X_vis, y, test_size=0.3, stratify=y, random_state=42
)

# Visual model (used for decision boundary plotting)
vis_model = DecisionTreeClassifier(random_state=42)
vis_model.fit(X_train_vis, y_train)

plot_decision_boundary(X_train_vis, y_train, vis_model, "Decision Tree - Training Set", label_encoder=le)
plot_decision_boundary(X_test_vis, y_test, vis_model, "Decision Tree - Test Set", label_encoder=le)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
models = [
    ("Decision Tree", DecisionTreeClassifier(random_state=42)),
    ("K-Nearest Neighbors", KNeighborsClassifier(n_neighbors=5)),
    ("SVM (RBF kernel)", SVC(kernel='rbf', gamma='scale', C=1.0, probability=True)),
    ("Logistic Regression", LogisticRegression(multi_class='auto', solver='lbfgs', max_iter=200))
]
```

```python
def compare_classifiers(models, X_train, y_train, X_test, y_test, label_encoder, cmap_colors=('pink', 'skyblue', 'gray')):
    X1 = X_train.iloc[:, 0].values
    X2 = X_train.iloc[:, 1].values

    x1_min, x1_max = X1.min() - 1, X1.max() + 1
    x2_min, x2_max = X2.min() - 1, X2.max() + 1
    xx1, xx2 = np.meshgrid(
        np.arange(x1_min, x1_max, 0.01),
        np.arange(x2_min, x2_max, 0.01)
    )

    fig, axes = plt.subplots(1, len(models), figsize=(6 * len(models), 5))

    for idx, (name, model) in enumerate(models):
        model.fit(X_train, y_train)
        test_acc = model.score(X_test, y_test)

        grid_df = pd.DataFrame(np.array([xx1.ravel(), xx2.ravel()]).T, columns=X_train.columns)
        Z = model.predict(grid_df).reshape(xx1.shape)

        ax = axes[idx]
        ax.contourf(xx1, xx2, Z, alpha=0.5, cmap=ListedColormap(cmap_colors))
        ax.set_xlim(xx1.min(), xx1.max())
        ax.set_ylim(xx2.min(), xx2.max())

        for i, cls in enumerate(np.unique(y_train)):
            class_label = label_encoder.inverse_transform([cls])[0]
            ax.scatter(
                X1[y_train == cls], X2[y_train == cls],
                c=[ListedColormap(cmap_colors)(i)],
                label=class_label,
                edgecolor='k', s=60
            )

        ax.set_title(f"{name}\nTest Accuracy: {test_acc:.2f}")
        ax.set_xlabel(X_train.columns[0])
        ax.set_ylabel(X_train.columns[1])
        ax.legend(title="Species")
        ax.grid(True)

    plt.suptitle("Classifier Comparison (Decision Boundaries + Test Accuracy)", fontsize=18)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
compare_classifiers(models, X_train_vis, y_train, X_test_vis, y_test, label_encoder=le)
```
