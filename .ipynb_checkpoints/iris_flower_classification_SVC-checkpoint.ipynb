# üå∏ Scalable SVC Pipeline with Grid Search & Class-Based Design

## üìå Project Objective
In this notebook, we build a reusable and scalable machine learning workflow using Support Vector Classification (SVC) applied to the Iris dataset.

### Key goals:
- Use `Pipeline` to streamline preprocessing + model training
- Use `GridSearchCV` to tune hyperparameters (`C`, `gamma`, `kernel`)
- Wrap everything into a custom Python class (`SVCClassifierPipeline`)
- Visualize decision boundaries and model performance
- Save and reuse the best-performing model

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import joblib
import os
```

## üß™ 1. Data Loading & Exploration

```python
iris = load_iris(as_frame=True)
df = iris.frame
df.head()
```

## üßº 2. Feature Engineering & Preprocessing

```python
le = LabelEncoder()
df['target'] = le.fit_transform(df['target'])

X = df[['petal length (cm)', 'petal width (cm)']]
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)
```

## ‚öôÔ∏è 3. SVCClassifierPipeline Class

```python
class SVCClassifierPipeline:
    def __init__(self, model_name="SVC"):
        self.model_name = model_name
        self.pipeline = None
        self.best_params_ = None
        self.grid_search = None
        self.best_model = None

    def build_pipeline(self):
        """Creates a pipeline with preprocessing and SVC."""
        self.pipeline = Pipeline([
            ("scaler", StandardScaler()),
            ("svc", SVC())
        ])

    def tune_hyperparameters(self, X_train, y_train, param_grid, cv=5, scoring='accuracy'):
        """Uses GridSearchCV to find beset parameters."""
        self.build_pipeline()
        self.grid_search = GridSearchCV(self.pipeline, param_grid=param_grid, cv=cv, scoring=scoring)
        self.grid_search.fit(X_train, y_train)
        self.pipeline = self.grid_search.best_estimator_
        self.best_params_ = self.grid_search.best_params_
        self.best_model = self.grid_search.best_estimator_

    def train(self, X_train, y_train):
        """Trains the current pipeline (if skipping grid search)."""
        if self.pipeline is None:
            self.build_pipeline()
        self.pipeline.fit(X_train, y_train)

    def evaluate(self, X_test, y_test, label_encoder=None, show_confusion_matrix=True, show_class_report=True, save_plots=False, output_dir="plots"):
        """Prints evaluation results and optionally saves visualizations."""
        y_pred = self.pipeline.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        print(f"\n Evaluation for {self.model_name}")
        print(f"Accuracy: {accuracy:.2f}")
        print("\n Classification Report:")
        print(classification_report(y_test, y_pred, target_names=[str(cls) for cls in label_encoder.classes_] if label_encoder else None))

        if not os.path.exists(output_dir) and save_plots:
            os.makedirs(output_dir)

        if show_confusion_matrix or save_plots:
            cm = confusion_matrix(y_test, y_pred)
            classes = label_encoder.classes_ if label_encoder else np.unique(y_test)
            cm_df = pd.DataFrame(cm, index=classes, columns=classes)

            plt.figure(figsize=(6, 5))
            sns.heatmap(cm_df, annot=True, fmt="d", cmap="viridis", cbar=False, linewidths=0.5)
            plt.title(f"Confusion Matrix = {self.model_name}")
            plt.ylabel("Actual")
            plt.xlabel("Predicted")
            plt.tight_layout()

            if save_plots:
                cm_path = os.path.join(output_dir, f"{self.model_name}_confusion_matrix.png")
                plt.savefig(cm_path, dpi=300)
                print(f"Confusion matrix saved to {cm_path}")
            if show_confusion_matrix:
                plt.show()
            else:
                plt.close()

        if show_class_report or save_plots:
            report_dict = classification_report(
                y_test, y_pred, target_names=label_encoder.classes_ if label_encoder else None, output_dict=True
            )

            # Remove 'accuracy', 'macro avg', and 'weighted avg' for simplicity
            report_df = pd.DataFrame(report_dict).drop(columns=['accuracy', 'macro avg', 'weighted avg'], errors='ignore').T

            plt.figure(figsize=(6,4))
            sns.heatmap(report_df, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=0.5)
            plt.title(f"Classification Report - {self.model_name}")
            plt.ylabel("Class")
            plt.xlabel("Metric")
            plt.tight_layout()

            if save_plots:
                cr_path = os.path.join(output_dir, f"{self.model_name}_classification_report.png")
                plt.savefig(cr_path, dpi=300)
                print(f"Classification report saved to {cr_path}")
            if show_class_report:
                plt.show()
            else:
                plt.close()

    def predict(self, X):
        return self.pipeline.predict(X)

    def save_model(self, path="svc_model.pk1"):
        joblib.dump(self.pipeline, path)
        print(f"Model saved to {path}")

    def load_model(self, path="svc_model.pk1"):
        self.pipeline = joblib.load(path)
        print(f"Model loaded from {path}")
```

## üîç 4. Hyperparameter Tuning

```python
param_grid = {
    'svc__C': [0.1, 1, 2, 10],
    'svc__gamma': ['scale', 0.01, 0.05, 0.1, 1],
    'svc__kernel': ['rbf']
}

svc_pipeline = SVCClassifierPipeline()
svc_pipeline.tune_hyperparameters(X_train, y_train, param_grid)
svc_pipeline.best_model.probability = True
print("Best Params:", svc_pipeline.best_params_)
```

## üìä 5. Grid Search Visualization

```python
def plot_grid_scores(grid_search, kernel_filter="rbf", score_metric='mean_test_score'):
    """Visualizes GridSearchCV results as a heatmap for a specific kernel (e.g., 'rbf')."""
    results = pd.DataFrame(grid_search.cv_results_)
    
    # Filter rows for just one kernel (e.g., 'rbf')
    filtered = results[results['param_svc__kernel'] == kernel_filter]
    
    try:
        pivot = filtered.pivot(index='param_svc__gamma', columns='param_svc__C', values=score_metric)
        sns.heatmap(pivot, annot=True, fmt=".3f", cmap="viridis")
        plt.title(f"GridSearchCV Mean Test Score ({kernel_filter.upper()} Kernel)")
        plt.xlabel("C")
        plt.ylabel("Gamma")
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print("‚ö†Ô∏è Could not create pivot table:", e)

plot_grid_scores(svc_pipeline.grid_search, kernel_filter="rbf")
```

## üéØ 6. Model Evaluation

```python
# Evaluate on set
svc_pipeline.evaluate(X_test, y_test, label_encoder=le, save_plots=True)
```

## üíæ 7. Save Best Model

```python
svc_pipeline.save_model('svc_best_model.pkl')
```
